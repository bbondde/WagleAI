```python
import os
import cv2
import glob
from xml.etree.ElementTree import parse
```

```python
#####################
## path initialize ##
#####################

print(os.getcwd())

#Pedestrian_data_path = os.getcwd() + '/data/Pedestrian'
Pedestrian_data_path = os.getcwd() + '/data/temp_pedestrian'
pedestrian_folder_list = os.listdir(Pedestrian_data_path)

pedestrian_folder_list.sort()
print(pedestrian_folder_list)
```

```python
#####################
## removing images ##
#####################

for ped_fol in pedestrian_folder_list:
    Bbox_data_path = Pedestrian_data_path + "/" + ped_fol
    xml_file_folder_list = os.listdir(Bbox_data_path)
    xml_file_folder_list.sort()
    #print(xml_file_folder_list)
    
    for xml_fol in xml_file_folder_list:
        Xml_file_path = Bbox_data_path + "/" + xml_fol
        xml_file_list= [a for a in os.listdir(Xml_file_path) if a.endswith(r".xml")]
        print(xml_fol, xml_file_list)
        
        xml_tree = parse(Xml_file_path + "/" + xml_file_list[0])
        xml_root = xml_tree.getroot()
        
        image_list = []
        
        for images in xml_root.iter("image"):
            label_list = []
            for boxes in images.iter("box"):
                label_list.append(boxes.attrib["label"])
            
            if "person" in label_list:
                image_list.append(images.attrib["name"])
                
        #print(image_list)
                
        image_file_list = [a for a in os.listdir(Xml_file_path) if a.endswith(r".jpg")]
        for file in image_file_list:
            if file not in image_list:
                os.remove(Xml_file_path + "/" + file)
```

```python
###################
## make xml file ##
###################

from xml.etree.ElementTree import Element, SubElement, ElementTree

for ped_fol in pedestrian_folder_list:
    Bbox_data_path = Pedestrian_data_path + "/" + ped_fol
    xml_file_folder_list = os.listdir(Bbox_data_path)
    xml_file_folder_list.sort()
    #print(xml_file_folder_list)
    
    for xml_fol in xml_file_folder_list:
        Xml_file_path = Bbox_data_path + "/" + xml_fol
        xml_file_list= [a for a in os.listdir(Xml_file_path) if a.endswith(r".xml")]
        image_file_list = [a for a in os.listdir(Xml_file_path) if a.endswith(r".jpg")]
        print(xml_fol, xml_file_list)
        
        xml_tree = parse(Xml_file_path + "/" + xml_file_list[0])
        xml_root = xml_tree.getroot()
        
        image_list = []
        
        for images in xml_root.iter("image"):
            label_list = []
            for boxes in images.iter("box"):
                label_list.append(boxes.attrib["label"])
            
            if "person" in label_list:
                image_list.append(images.attrib["name"])
                
        #print(image_list)
        

        for images in xml_root.iter("image"):
            
            label_list = []
            
            
            
            #if images.attrib["name"] in image_list:
            if images.attrib["name"] in image_file_list:    
            #if images.attrib["name"] in ["MP_KSC_000001.jpg"]:
                width = images.attrib["width"]
                height = images.attrib["height"]                
                
                for boxes in images.iter("box"):
                    if boxes.attrib["label"] == "person":
                        label_list.append([width, height, boxes.attrib["label"],
                                           boxes.attrib["xtl"], boxes.attrib["ytl"],
                                           boxes.attrib["xbr"], boxes.attrib["ybr"]])
                       
                
            else :
                continue
            
            root = Element('annotation')
            SubElement(root, 'folder').text = Xml_file_path
            SubElement(root, 'filename').text = images.attrib["name"]
            SubElement(root, 'path').text = Xml_file_path + '/' + images.attrib["name"]
            source = SubElement(root, 'source')
            SubElement(source, 'database').text = 'Unknown'
            
            size = SubElement(root, 'size')
            SubElement(size, 'width').text = str(label_list[0][0])
            SubElement(size, 'height').text = str(label_list[0][1])
            SubElement(size, 'depth').text = '3'
            
            SubElement(root, 'segmented').text = '0'
            
            for label in label_list:
                obj = SubElement(root, 'object')
                SubElement(obj, 'name').text = label[2]
                SubElement(obj, 'pose').text = 'Unspecified'
                SubElement(obj, 'truncated').text = '0'
                SubElement(obj, 'difficult').text = '0'
                
                box = SubElement(obj, 'bndbox')
                SubElement(box, 'xmin').text = label[3]
                SubElement(box, 'ymin').text = label[4]
                SubElement(box, 'xmax').text = label[5]
                SubElement(box, 'ymax').text = label[6]
                
            tree = ElementTree(root)
            tree.write(Xml_file_path + "/" + images.attrib["name"][:-4] + '.xml')
                     
                    
        
        
```
